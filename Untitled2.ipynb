```python
def normalize(x):
    z = x.copy()
    z -= z.min()
    z /= z.max()
    return z
```

```python
%matplotlib notebook
plt.figure()

plt.hist(normalize(normalize(df2['toxic'].values)), bins=30, color='b')
plt.hist(normalize(normalize(df1['toxic'].values)), bins=30, color='r')
plt.show()
```

```python
%matplotlib notebook
plt.figure()
plt.hist(normalize(normalize(df1['toxic'].values)), bins=30, color='r')
plt.hist(normalize(normalize(df2['toxic'].values)), bins=30, color='b')

plt.show()
```

```python
# %matplotlib notebook
plt.figure()
plt.hist(df2['toxic'] - df1['toxic'], bins=30, color='r')
plt.hist(normalize(df2['toxic']) - normalize(df1['toxic']), bins=30, color='b')

plt.show()
```
